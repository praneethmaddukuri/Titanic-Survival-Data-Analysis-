# Question 1: Gender & Class Impact on Survival
- How did gender and passenger class impact survival rates?
- Provide survival rate comparisons using tables and visualizations.

```python
plt.figure(figsize=(10, 5))
sns.barplot(x="Pclass", y="Survived", hue="Sex", data=df)
plt.title("Survival Rate by Gender and Passenger Class")
plt.show()
```

```python
survival_gender_class = df.groupby(['Pclass', 'Sex'])['Survived'].mean()
display(survival_gender_class)
```

```python
print("1.Women had a significantly higher survival rate across all classes, especially in 1st and 2nd class. Men in lower classes had the lowest survival rates.")
```

# Question 2: Family Size & Survival
- How does traveling alone compare to traveling with family in terms of survival chances?
- Categorize passengers based on family size and analyze survival rates.0

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
```

```python
def family_category(size):
    if size == 1:
        return 'Alone'
    elif size <= 4:
        return 'Small Family'
    else:
        return 'Large Family'

df['FamilyCategory'] = df['FamilySize'].apply(family_category)
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x="FamilyCategory", y="Survived", data=df, palette="Set1")
plt.title("Survival Rate by Family Size")
plt.show()
```

```python
print("2. People traveling alone had the lowest survival rates, while small families had a better chance of survival. Large families had slightly lower survival rates than small families but higher than those traveling alone.")
```

# Question 3: Embarkation Port & Survival
- Did the port where passengers boarded affect their survival rates?
- Compare survival rates among passengers who embarked from different ports (C, S, Q).

```python
plt.figure(figsize=(8, 5))
sns.barplot(x="Embarked", y="Survived", data=df, palette={"C": "blue", "Q": "green", "S": "red"})
plt.title("Survival Rate by Embarkation Port")
plt.show()
```

```python
print("3. Passengers who embarked from port C (blue) had the highest survival rates, followed by those from Q (green). Passengers from S (red) had the lowest survival rates.")
```

# Question 4: Hidden Pattern - Age Distribution
- Identify another hidden pattern in the dataset that affects survival rates.
- Justify your findings with data and graphs.

```python
plt.figure(figsize=(6, 4))
sns.histplot(df[df['Survived'] == 1]['Age'].dropna(), bins=30, kde=True, color='green', label='Survived')
sns.histplot(df[df['Survived'] == 0]['Age'].dropna(), bins=30, kde=True, color='red', label='Not Survived')
plt.title("Age Distribution of Survivors vs Non-Survivors")
plt.legend()
plt.show()
```

```python
print("4. Younger passengers had a better chance of survival compared to older passengers, especially children.")
```
